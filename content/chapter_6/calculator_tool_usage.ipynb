```python
# Demonstarting how to use langchain.debug to print out the intermediate results with stack traces:
import langchain

langchain.debug = True
```

```python
from langchain import (
    LLMMathChain,
)
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType
from langchain.chat_models import ChatOpenAI
```

```python
llm = ChatOpenAI(temperature=0)
llm_math_chain = LLMMathChain.from_llm(llm=llm, verbose=True)
tools = [
    Tool(
        name="Calculator",
        func=llm_math_chain.run,
        description="useful for when you need to answer questions about math",
        return_direct=True,
    ),
]
```

```python
agent = initialize_agent(tools, llm, agent=AgentType.OPENAI_FUNCTIONS, verbose=True)
```

```python
result = agent.run("What is 5 + 5? Return only the number.")
print(result)
```

```python
llm = ChatOpenAI(temperature=0)
llm_math_chain = LLMMathChain.from_llm(llm=llm, verbose=True)


def count_characters_in_string(string: str):
    return len(string)


tools = [
    Tool(
        func=llm_math_chain.run,
        name="Calculator",
        description="useful for when you need to answer questions about math",
    ),
    Tool(
        func=count_characters_in_string,
        name="Count Characters",
        description="useful for when you need to count the number of characters in a string",
    ),
]
agent = initialize_agent(
    tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True
)

result = agent.run(
    """
    Task: What's the character length of the following sentence: The quick brown fox jumps over the lazy dog?
    After you have found the answer, then square it.
    """
)
print(result)
```
